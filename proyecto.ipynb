#### Elaboración de función de entrenamiento

```python
def entrenamiento(set_entre, epochs = 5000,  imprimir_error_cada=10, learning_rate=0.001):
    x = pd.DataFrame(set_entre.iloc[:,0]).values
    y = pd.DataFrame(set_entre.iloc[:,1]).values
    #inicialización de parametros b0 y b1
    b0 = 0.1
    b1 = 0.2
    x1s = np.ones_like(x).reshape(-1,1)
    x=np.reshape(x, (-1,1))
    matriz = np.hstack([x, x1s])
    betas = [b0, b1]
    errores = []
    df_errores = pd.DataFrame(columns=["Epoch", "Error", "B0", "B1"])
    for i in range(epochs):
        y_estimada = np.reshape(np.dot(matriz, betas), (-1,1))
        error = np.mean(np.power(y-y_estimada,2))
        b0 = b0 - learning_rate*b0
        b1 = b1 - learning_rate*b1
        betas = [b0, b1]
        errores.append(error)
        df_errores = pd.concat([df_errores, pd.DataFrame({"Epoch": [i], "Error":error, "B0":b0,"B1":b1})])
    print(df_errores['Epoch'], df_errores['Error'])
    df_errores['Error'].plot()
    #return df_errores, df_errores.iloc[-1].to_dict()
```

```python
entrenamiento(set_entrenamiento)
#display(modelo)
#modelo.plot()
```
