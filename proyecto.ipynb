#### Elaboración de función de entrenamiento

```python
def entrenamiento(set_entre, epochs = 5000,  imprimir_error_cada=100, learning_rate=0.001):
    x = pd.DataFrame(set_entre.iloc[:,0]).values
    y = pd.DataFrame(set_entre.iloc[:,1]).values
    #inicialización de parametros b0 y b1
    b0 = 0.1
    b1 = 0.2
    x1s = np.ones_like(x).reshape(-1,1)
    x=np.reshape(x, (-1,1))
    matriz = np.hstack([x, x1s])
    betas = [b0, b1]
    errores = []
    df_errores = pd.DataFrame(columns=["Epoch", "Error", "B0", "B1"])
    for i in range(epochs):
        y_estimada = np.reshape(np.dot(matriz, betas), (-1,1))
        error = np.mean(np.power(y-y_estimada,2))
        #Error promedio se utiliza para para mostrar el error promedio de acuerdo al parametro "imprimir_error_cada"
        error_promedio = np.divide(np.mean(np.power(error,2)),2)
        b0 = b0 - learning_rate*b0
        b1 = b1 - learning_rate*b1
        betas = [b0, b1]
        errores.append(error)
        df_errores = pd.concat([df_errores, pd.DataFrame({"Epoch": [i], "Error":error, "B0":b0,"B1":b1})])
        if i % imprimir_error_cada == 0:
            print(f" Epoch: {i}, Error Promedio: {error_promedio}, b0: {b0}, b1: {b1} ")
    return df_errores, df_errores.iloc[-1].to_dict()
```

```python
modelo, betas = entrenamiento(set_entrenamiento)
```

```python
def grafica_errores(errores):
    errores.plot(x="Epoch", y="Error")
    plt.ylabel("Error")
    plt.title("Gráfica de errores - Función de entrenamiento")
```

```python
histograma_errores = grafica_errores(modelo)
histograma_errores
```

#### Elaboración de función de visualización

```python
def visualizacion(modelo, set_evaluacion):
    x = pd.DataFrame(set_evaluacion.iloc[:,0])
    x1s = np.ones_like(x).reshape(-1,1)
    x=np.reshape(x, (-1,1))
    matriz = np.hstack([x, x1s])
    betas = [modelo['B1'], modelo['B0']]
    y_estimada = np.dot(matriz, betas)
    return pd.DataFrame({'X': set_evaluacion.iloc[:,1], 'Y estimada': y_estimada})
```
