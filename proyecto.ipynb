```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
#Carga del data set
data_set = np.load('./proyecto_data/proyecto_training_data.npy', 'r')
df = pd.DataFrame(data_set, columns = ['PrecioVenta', 'Calificación', 'ÁreaPP', 'Habitaciones', 'AñoConstrucción', 'Frente'])
df.describe()
```

```python
#Visualización de las dimensiones del data set
filas, columnas = data_set.shape

#utilizando slicing para separar en 80% y 20% el data set
tam_set_entrenamiento = int(filas*0.8)
tam_set_prueba = int(filas-tam_set_entrenamiento)
print(f"Tamaño del set de entrenamiento: {tam_set_entrenamiento}")
print(f"Tamaño del set de prueba: {tam_set_prueba}")

#almacenanndo los dos sets en diferntes porciones
set_entrenamiento = df.sample(frac=0.8)
set_prueba = df.drop(set_entrenamiento.index)
```

#### Análisis exploratorio del set de entrenamiento

```python
#Análisis columan 1 PrecioVenta
media_PV = set_entrenamiento['PrecioVenta'].mean()
max_PV = set_entrenamiento['PrecioVenta'].max()
min_PV = set_entrenamiento['PrecioVenta'].min()
std_PV = set_entrenamiento['PrecioVenta'].std()

#Análisis columan 2 Calificación
media_PV = set_entrenamiento['Calificación'].mean()
max_PV = set_entrenamiento['Calificación'].max()
min_PV = set_entrenamiento['Calificación'].min()
std_PV = set_entrenamiento['Calificación'].std()

#Análisis columan 3 ÁreaPP
media_PV = set_entrenamiento['ÁreaPP'].mean()
max_PV = set_entrenamiento['ÁreaPP'].max()
min_PV = set_entrenamiento['ÁreaPP'].min()
std_PV = set_entrenamiento['ÁreaPP'].std()

#Análisis columan 4 Habitaciones
media_PV = set_entrenamiento['Habitaciones'].mean()
max_PV = set_entrenamiento['Habitaciones'].max()
min_PV = set_entrenamiento['Habitaciones'].min()
std_PV = set_entrenamiento['Habitaciones'].std()

#Análisis columan 5 AñoConstrucción
media_PV = set_entrenamiento['AñoConstrucción'].mean()
max_PV = set_entrenamiento['AñoConstrucción'].max()
min_PV = set_entrenamiento['AñoConstrucción'].min()
std_PV = set_entrenamiento['AñoConstrucción'].std()

#Análisis columan 5 Frente
media_PV = set_entrenamiento['Frente'].mean()
max_PV = set_entrenamiento['Frente'].max()
min_PV = set_entrenamiento['Frente'].min()
std_PV = set_entrenamiento['Frente'].std()
```

### Cálculo y grágica de la correlación

```python
correlacion = set_entrenamiento.corr()
correlacion
```

```python
fig, axs = plt.subplots(len(set_entrenamiento.columns), len(set_entrenamiento.columns), figsize=(30, 30))
fig.suptitle("Diagramas de dispersión", fontsize=32)
for x in range(len(set_entrenamiento.columns)):
    for y in range(len(set_entrenamiento.columns)):
        axs[x,y].scatter(set_entrenamiento[set_entrenamiento.columns[x]], set_entrenamiento[correlacion.columns[y]])
        axs[x,y].set_xlabel(set_entrenamiento.columns[y])
        val_corr = str(set_entrenamiento.corr().iloc[x,y])
        titulo = "Val. Correlación " + val_corr
        axs[x,y].set_title(titulo)
        if y == 0:
            axs[x,y].set_ylabel(set_entrenamiento.columns[x])
```

#### Variables con mayor potencial predictivo

El valor de correlación más alto obtenido es 0.827809 de la relación entre el Precio de Venta y la Calificación


```python
valor1 = plt.plot(set_entrenamiento['PrecioVenta'], set_entrenamiento['Calificación'], "bo")
valor1
```
